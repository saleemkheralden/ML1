```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataset = load_wine()
df = pd.DataFrame(data=dataset['data'], columns=dataset['feature_names'])
df = df.assign(target=pd.Series(dataset['target']).values)
```

```python
df = df[['alcohol', 'magnesium', 'target']]
df = df[df.target != 0]
```

```python
train_df, val_df = train_test_split(df, test_size=30, random_state=3)
```

Question 1:

```python
colors = np.array(['red', 'blue'])
plt.scatter(train_df['alcohol'], train_df['magnesium'], label='train set', c=colors[train_df['target'] - 1])
plt.xlabel('alcohol concentration')
plt.ylabel('magnesium concentration')
plt.legend()
# plt.scatter(val_df['alcohol'], val_df['magnesium'], label='test set', color='red')
```
